```python
import numpy as np
from scipy.stats import wilcoxon, ttest_rel
import matplotlib.pyplot as plt
import pandas as pd

lstm_64_250 = pd.read_csv("lstm_64_250.csv")
lstm_64_250_acc = lstm_64_250["accuracy"]

lstm_128_500 = pd.read_csv("lstm_128_500.csv")
lstm_128_500_acc = lstm_128_500["accuracy"]

plt.scatter(lstm_64_250_acc, np.zeros(len(lstm_64_250_acc)), c = 'r') # plot first line
plt.scatter(lstm_128_500_acc, np.zeros(len(lstm_128_500_acc)) + 1, c = 'b') # plot first line


datax = [lstm_64_250_acc, lstm_128_500_acc]
datay = [np.zeros(len(lstm_64_250_acc)), (np.zeros(len(lstm_128_500_acc)) + 1)]
_ = plt.plot(datax, datay, 'yo', linestyle="--", alpha = 0.2)

# stat, p = wilcoxon(lstm_64_250_acc, lstm_128_500_acc, mode="approx")
wilcoxon_stat, wilcoxon_pval = wilcoxon(lstm_64_250_acc, lstm_128_500_acc, alternative='two-sided', zero_method='wilcox')
t_stat, ttest_pval = ttest_rel(lstm_64_250_acc, lstm_128_500_acc)

print('Wilcoxon: p-value=%.3f' % (wilcoxon_pval))
print('T-test: p-value=%.3f' % (ttest_pval))

if wilcoxon_pval > 0.05:
  print('Wilcoxon: Probably the same distribution')
else:
  print('Wilcoxon: Probably different distributions')

if ttest_pval > 0.05:
  print('T-test: Probably the same distribution')
else:
  print('T-test: Probably different distributions')
```

```python
import seaborn as sns
# display test results
print("Wilcoxon Signed-Rank Test:")
#print(f"Statistic: {wilcoxon_stat}")
print(f"p-value: {wilcoxon_pval}")

print(lstm_64_250_acc.to_numpy())
print("-----------------")
print(lstm_128_500_acc.to_numpy())

m1_name = "batch 64 text 250"
m2_name = "batch 128 text 500"

dat = pd.DataFrame({
  "Model" : [m1_name, m1_name,m1_name,m1_name,m1_name,m2_name,m2_name,m2_name,m2_name,m2_name,], 
  "Performance" : [0.871, 0.8667, 0.868, 0.8603, 0.865, 0.8621, 0.8484, 0.854, 0.8673, 0.8701]
})

sns.boxplot(data=dat, x='Model', y='Performance')
sns.stripplot(data=dat, x='Model', y='Performance', color='black', size=5)
```
